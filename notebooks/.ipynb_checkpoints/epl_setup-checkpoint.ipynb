
    I(gl::GaussianLaser)

returns the beam Intensity: $I(\rho, z) = I_0 ( W_0 / W(z))^2 \exp{-2 \rho^2/W^2(z)} $

#### Fields
- `gl::GaussianLaser`       : A gaussian Laser 


```julia
fI = LaserLab.I(g2epl375) 
```

```julia
wl=-5.0:0.01:5.0
plot(wl, fI.(0.0*μm, wl*μm)/(mW * cm^-2), label="I(0,z)")
ylims!((0.0,1.10e+7))
xlabel!("z (μm)")
ylabel!("I(0,z)")
```

```julia
wl=-1.0:0.01:1.0
plot(wl, fI.(wl*μm, 0.0*μm )/(mW * cm^-2), label="I(ρ,0)")
ylims!((0.0,1.1e+7))
xlabel!("ρ (μm)")
ylabel!("I(ρ,0)")
```

```julia
zl=-5.0:0.01:5.0
rl=-1.0:0.01:1.0
zs = [fI.(x*μm, y*μm )/(mW * cm^-2) for y in zl, x in rl]
surface(rl, zl, zs)
```

```julia
wl=-1.0:0.01:1.0
p1 = plot(wl, fI.(wl*μm, 0.0*μm )/(mW * cm^-2), label="I(ρ,0)")
p2 = plot!(p1, wl, fI.(wl*μm, 0.5*μm )/(mW * cm^-2), label="I(ρ,0.5)")

ylims!((0.0,1.1e+7))
xlabel!("ρ (μm)")
ylabel!("I(ρ,0)")
```

```julia
wl=-1.0:0.01:1.0
p1 = plot(wl, fI.(wl*μm, 0.0*μm )/(mW * cm^-2), label="I(ρ,0)")
p2 = plot!(p1, wl, fI.(wl*μm, 0.5*μm )/(mW * cm^-2), label="I(ρ,0.5)")
p3 = plot!(p2, wl, fI.(wl*μm, 1.0*μm )/(mW * cm^-2), label="I(ρ,1.0)")
ylims!((0.0,1.1e+7))
xlabel!("ρ (μm)")
ylabel!("I(ρ,0)")
```
